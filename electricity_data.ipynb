```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'C:/Users/crick/OneDrive/Documents/GitHub/COVID19PollutionProject/'
file = 'sales_revenue.xlsx'
```

```python
#read in data
electricity_data = pd.read_excel(file_path+file)
```

```python
#preview data
electricity_data.head()
```

```python
#subset data frame into one for residentail consumption
#and one for commercial consumption
residential_df = electricity_data[['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 2', 'RESIDENTIAL', 'Unnamed: 5', 
                                  'Unnamed: 6', 'Unnamed: 7']]
commercial_df = electricity_data[['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 2', 'COMMERCIAL', 'Unnamed: 9', 
                                  'Unnamed: 10', 'Unnamed: 11']]
```

```python
#preview
residential_df.head()
```

```python
commercial_df.head()
```

```python
#rename columns
residential_df = residential_df.rename(columns={'Unnamed: 0': 'Year', 'Unnamed: 1': 'Month', 'Unnamed: 2': 'State',
                                               'RESIDENTIAL': 'Revenue Thousand Dollars', 'Unnamed: 5': 'Sales Megawatthours', 
                                               'Unnamed: 6': 'Customers', 'Unnamed: 7': 'Price Cents/kWh'})
commercial_df = commercial_df.rename(columns={'Unnamed: 0': 'Year', 'Unnamed: 1': 'Month', 'Unnamed: 2': 'State',
                                              'COMMERCIAL': 'Revenue Thousand Dollars', 'Unnamed: 9': 'Sales Megawatthours', 
                                              'Unnamed: 10': 'Customers', 'Unnamed: 11': 'Price Cents/kWh'})
```

```python
residential_df.head()
```

```python
commercial_df.head()
```

```python
#drop old column headers
residential_df = residential_df.drop([0,1])
residential_df.head()
```

```python
commercial_df = commercial_df.drop([0,1])
commercial_df.head()
```

```python
#filter data frames to only have DC as the State, and create a datetime column
residential_dc_df = residential_df.loc[residential_df['State'] == 'DC']
residential_dc_df['Date'] = pd.to_datetime(residential_dc_df[['Year', 'Month']].assign(DAY=1))
residential_dc_df.head()
```

```python
commercial_dc_df = commercial_df.loc[commercial_df['State'] == 'DC']
commercial_dc_df['Date'] = pd.to_datetime(commercial_dc_df[['Year', 'Month']].assign(DAY=1))
commercial_dc_df.head()
```

```python
#make one useful dataframe
electricity_df = residential_dc_df[['Date', 'Sales Megawatthours']].merge(commercial_dc_df[['Date', 'Sales Megawatthours']],
                                                                          on = 'Date', suffixes=(' Residential', ' Commercial'))
electricity_df.tail()
```

```python
#filter out NaNs that occur in 2005
electricity_df = electricity_df.loc[electricity_df['Date']>='01-01-2006']
```

```python
#plot data together to compare visually 
plt.figure() 

electricity_df.plot('Date', ['Sales Megawatthours Residential', 'Sales Megawatthours Commercial'], 
         label=['Residential', 'Commercial'])
```

```python
#plot on a shorter timeframe
plt.figure() 

electricity_df.loc[electricity_df['Date'] >= '01-01-2015'].plot('Date', 
                                                                ['Sales Megawatthours Residential', 'Sales Megawatthours Commercial'], 
                                                                label=['Residential', 'Commercial'])
```

```python
#plot separately on the shorter time frame
plt.figure() 
#plot both data sets
electricity_df.loc[electricity_df['Date'] >= '01-01-2015'].plot('Date', 'Sales Megawatthours Residential')
electricity_df.loc[electricity_df['Date'] >= '01-01-2015'].plot('Date', 'Sales Megawatthours Commercial')
```
